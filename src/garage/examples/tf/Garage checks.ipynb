```python
def plot_random_track(filenumber): 
    rf_file = rf[rf['file']==filenumber]
    
    event_r = pd.read_hdf('~/gnnfiles/data/ntuple_PU200_numEvent1000/ntuple_PU200_event' + str(filenumber)+'.h5')
    random_pid = random.choice(rf_file.particle_id.values)
    print(random_pid)

    true_hits = event_r[event_r['particle_id']==random_pid]
    #print(true_hits)
    rf_hits = rf_file[rf_file['particle_id']==random_pid]
    #print(rf_hits.reward)
    
    plt.plot(true_hits.z, true_hits.r, "kx", label="True hits")
    plt.plot(rf_hits.z, rf_hits.r, "b*", label="Reinforcement predictions")
    plt.xlabel("z (cm)")
    plt.ylabel("r (cm)")
    #plt.legend() 
    
    
    plt.show() 
```
